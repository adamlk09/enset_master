```python
# ============================================
# NOTEBOOK 2: NETTOYAGE DES DONNÉES
# ============================================

# Cellule 1: Charger les données
import pandas as pd
import numpy as np
from pathlib import Path

df_sales = pd.read_excel(r'C:\Users\pc\Desktop\enset\python-project\data\Sales.xlsx')
print(f"Données chargées: {df_sales.shape}")

# Cellule 2: Nettoyer les données
print("\n=== NETTOYAGE DES DONNÉES ===")

# Supprimer les doublons
n_duplicates = df_sales.duplicated().sum()
df_sales = df_sales.drop_duplicates()
print(f"✓ {n_duplicates} doublons supprimés")


print("\nValeurs manquantes avant :")
print(df_sales.isnull().sum())

colonnes_critiques = [
    'OrderDate',
    'Order Quantity',
    'Unit Selling Price',
    'Unit Cost'
]

df_sales.dropna(subset=colonnes_critiques, inplace=True)

print("✓ Lignes avec valeurs manquantes critiques supprimées")
df_sales['OrderDate'] = pd.to_datetime(df_sales['OrderDate'], errors='coerce')
df_sales['Order Quantity'] = pd.to_numeric(df_sales['Order Quantity'], errors='coerce')
df_sales['Unit Selling Price'] = pd.to_numeric(df_sales['Unit Selling Price'], errors='coerce')
df_sales['Unit Cost'] = pd.to_numeric(df_sales['Unit Cost'], errors='coerce')

print("✓ Types de données convertis")
df_sales['Sales'] = df_sales['Order Quantity'] * df_sales['Unit Selling Price']
df_sales['Cost'] = df_sales['Order Quantity'] * df_sales['Unit Cost']
df_sales['Profit'] = df_sales['Sales'] - df_sales['Cost']

df_sales['Profit Margin %'] = np.where(
    df_sales['Sales'] != 0,
    df_sales['Profit'] / df_sales['Sales'],
    0
)

print("✓ Colonnes Sales, Cost, Profit et Profit Margin créées")
df_sales['Year'] = df_sales['OrderDate'].dt.year
df_sales['Month'] = df_sales['OrderDate'].dt.month
df_sales['Quarter'] = df_sales['OrderDate'].dt.quarter
df_sales['Month Name'] = df_sales['OrderDate'].dt.strftime('%B')
df_sales['Day Name'] = df_sales['OrderDate'].dt.strftime('%A')

print("✓ Colonnes temporelles créées")
print("\n=== VALIDATION ===")
print(f"Forme finale : {df_sales.shape}")
print(f"Période : {df_sales['OrderDate'].min()} → {df_sales['OrderDate'].max()}")
print(f"Ventes totales : {df_sales['Sales'].sum():,.2f}")
print(f"Profit total : {df_sales['Profit'].sum():,.2f}")
print(f"Marge bénéficiaire moyenne : {df_sales['Profit Margin %'].mean():.2%}")
output_dir = Path('data/processed')
output_dir.mkdir(parents=True, exist_ok=True)

df_sales.to_csv(output_dir / 'sales_clean.csv', index=False)
df_sales.to_excel(output_dir / 'sales_clean.xlsx', index=False)

print("✓ Données nettoyées sauvegardées (CSV + Excel)")
```
